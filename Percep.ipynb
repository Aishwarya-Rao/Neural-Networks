```python
import numpy as np

class Perceptron():
    def __init__(self, learning_rate = 1, n_iterations = 5000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.bias = None
        self.weights = None
        
    
    def activation_function(self, net_input):
        # STEP ACTIVATION
        return np.where(net_input > 0.5, 1, 0)
    
    
    def fit(self, features, targets):
        n_examples, n_features = features.shape
        
        # change these to use different initialization scheme
        self.weights = np.random.uniform(size = n_features, low = -0.5, high = 0.5)
        self.bias = np.random.uniform(low = -0.5, high = 0.5)
        
        for _ in range(self.n_iterations):
            print("hwwhwar")
            for example_index, example_features in enumerate(features):
                print("mew:",example_index, example_features,self.weights,self.bias)
                net_input = np.dot(example_features, self.weights) + self.bias
                print(net_input)
                net_input = 1/(1+np.exp(-net_input)) 
                print(net_input)
                y_predicted = self.activation_function(net_input)
                self._update_weights(example_features, targets[example_index], y_predicted)
        
        
    def _update_weights(self, example_features, y_actual, y_predicted):
        error = y_actual - y_predicted  #if the value actual is more than predicted then the updated weight would be more than prev so that logically, when the inc weight is put in the eqn now the predicted value would be more than what it was, and continuing to tdo this there will be one point the value would be same which is what we want both actual and pred to be same. and the opposite if the pred is more than actual it decreases.
        print("errors:",y_actual,y_predicted,error)
        weight_correction = self.learning_rate * error
        self.weights = self.weights + weight_correction * example_features
        self.bias = self.bias + weight_correction
        
    def predict(self, features):
        net_input = np.dot(features, self.weights) + self.bias
        y_predicted = self.activation_function(net_input)
        return y_predicted 
```
